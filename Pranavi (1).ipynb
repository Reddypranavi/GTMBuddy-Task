```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import re

# Download NLTK resources if not already downloaded
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Load the dataset
file_path = "calls_dataset3.csv"  # Change to your actual CSV file path
data = pd.read_csv(file_path)

# Initialize lemmatizer and stopwords
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

# Function to preprocess text
def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    # Remove special characters, numbers, and punctuation
    text = re.sub(r'[^a-z\s]', '', text)
    # Tokenize the text
    tokens = word_tokenize(text)
    # Remove stopwords and lemmatize
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
    # Join tokens back into a string
    return ' '.join(tokens)

# Apply preprocessing to the Call_Text column
data['Processed_Text'] = data['Call_Text'].apply(preprocess_text)

# Save the preprocessed dataset to a new CSV
data.to_csv('calls_dataset_preprocessed.csv', index=False)

print("Preprocessing complete! The dataset with preprocessed text has been saved as 'calls_dataset_preprocessed.csv'.")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.multioutput import MultiOutputClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.metrics import classification_report
import numpy as np

# Check if the labels are lists of labels
print(y_train.head())  # This should show lists like ['label1', 'label2']

# Initialize the MultiLabelBinarizer to encode the labels
mlb = MultiLabelBinarizer()

# Transform the labels into multi-hot encoding format
y_train_encoded = mlb.fit_transform(y_train)
y_test_encoded = mlb.transform(y_test)

# List to store the indices of labels with only one class
labels_to_remove = []

# Check if any label has only one class
for i, label in enumerate(mlb.classes_):
    unique_classes = np.unique(y_train_encoded[:, i])
    print(f"Label '{label}' has unique classes: {unique_classes}")
    if len(unique_classes) == 1:  # If only one class exists for a label
        print(f"Label '{label}' only has one class. Marking this label for removal.")
        labels_to_remove.append(i)

# Remove labels that only have one class
y_train_encoded = np.delete(y_train_encoded, labels_to_remove, axis=1)
y_test_encoded = np.delete(y_test_encoded, labels_to_remove, axis=1)

# Update mlb.classes_ by removing the labels that only had one class
mlb.classes_ = np.delete(mlb.classes_, labels_to_remove)

# Initialize TfidfVectorizer
vectorizer = TfidfVectorizer()

# Fit and transform the training text data
X_train_vec = vectorizer.fit_transform(X_train)

# Transform the test text data using the same vectorizer
X_test_vec = vectorizer.transform(X_test)

# Check the shapes of the transformed data
print(f"X_train_vec shape: {X_train_vec.shape}")  # Should be (num_samples, num_features)
print(f"X_test_vec shape: {X_test_vec.shape}")

# Initialize the multi-output classifier with Logistic Regression (with class weights for imbalance handling)
multi_label_model = MultiOutputClassifier(LogisticRegression(class_weight='balanced'))

# Train the model on the training set
multi_label_model.fit(X_train_vec, y_train_encoded)

# Predict on the test set
y_pred_encoded = multi_label_model.predict(X_test_vec)

# Decode the predictions back to the original labels
y_pred_decoded = mlb.inverse_transform(y_pred_encoded)

# Print classification report
print("Classification Report:")
print(classification_report(y_test_encoded, y_pred_encoded, target_names=mlb.classes_))
```

```python
import json
import re
from collections import defaultdict

# Step 1: Load the Domain Knowledge
def load_domain_knowledge(file_path):
    """
    Load domain knowledge from a JSON file.
    :param file_path: Path to the JSON file.
    :return: Dictionary with domain knowledge.
    """
    with open(file_path, 'r') as file:
        data = json.load(file)
    return data

# Step 2: Perform Dictionary Lookup
def dictionary_lookup(text, domain_knowledge):
    """
    Extract entities from text based on a domain knowledge dictionary.
    :param text: Input text to search for entities.
    :param domain_knowledge: Dictionary containing domain-specific keywords.
    :return: Dictionary with extracted entities.
    """
    found_entities = defaultdict(list)

    # Check for matches in each category
    for category, keywords in domain_knowledge.items():
        for keyword in keywords:
            # Use regex for case-insensitive whole-word matching
            if re.search(r'\b' + re.escape(keyword) + r'\b', text, re.IGNORECASE):
                found_entities[category].append(keyword)

    return dict(found_entities)

# Step 3: Main Function to Test
def main():
    # Path to the domain knowledge JSON file
    domain_knowledge_file = 'domain_knowledge.json'

    # Example input text snippet
    text = "Our competitor, Competitor X, offers a product with high durability for $49.99."

    # Load the domain knowledge
    domain_knowledge = load_domain_knowledge(domain_knowledge_file)

    # Perform the dictionary lookup
    extracted_entities = dictionary_lookup(text, domain_knowledge)

    # Output the extracted entities
    print("Extracted Entities:")
    print(json.dumps(extracted_entities, indent=4))

if __name__ == "__main__":
    main()
```

```python
import json
import re
import spacy

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

def load_json(file_path):
    """Load the JSON file."""
    with open(file_path, 'r') as file:
        return json.load(file)

def dictionary_lookup(text, domain_knowledge):
    """Search for entities using a dictionary lookup."""
    found_entities = {"competitors": [], "features": [], "pricing": []}
    
    # Search for competitors
    for competitor in domain_knowledge.get("competitors", []):
        if re.search(r'\b' + re.escape(competitor) + r'\b', text, re.IGNORECASE):
            found_entities["competitors"].append(competitor)
    
    # Search for features
    for feature in domain_knowledge.get("features", []):
        if re.search(r'\b' + re.escape(feature) + r'\b', text, re.IGNORECASE):
            found_entities["features"].append(feature)
    
    # Search for pricing-related keywords
    for price_keyword in domain_knowledge.get("pricing", []):
        if re.search(r'\b' + re.escape(price_keyword) + r'\b', text, re.IGNORECASE):
            found_entities["pricing"].append(price_keyword)
    
    return found_entities

def ner_extraction_spacy(text):
    """Perform Named Entity Recognition using spaCy."""
    doc = nlp(text)
    entities = [(ent.text, ent.label_) for ent in doc.ents]
    return entities

def combine_extractions(dict_entities, ner_entities):
    """Combine results from dictionary lookup and NER."""
    combined = {**dict_entities}  # Start with dictionary entities
    combined["NER"] = list(set(ner_entities))  # Add unique NER entities
    return combined

# Main function
def main():
    domain_knowledge_file = "domain_knowledge.json"  # Path to your JSON file
    domain_knowledge = load_json(domain_knowledge_file)
    
    # Example text
    text = "BrandX provides affordable internet and excellent customer support."
    
    # Perform dictionary lookup and NER
    dict_entities = dictionary_lookup(text, domain_knowledge)
    ner_entities = [ent[0] for ent in ner_extraction_spacy(text)]  # Extract entity names only
    
    # Combine extractions
    final_entities = combine_extractions(dict_entities, ner_entities)
    print("Extracted Entities:", final_entities)

if __name__ == "__main__":
    main()
```

```python
# Function to output structured results
def output_entities(snippets, domain_knowledge):
    results = []
    for snippet in snippets:
        text = snippet["text"]

        # Dictionary lookup
        dict_entities = dictionary_lookup(text, domain_knowledge)

        # NER extraction
        ner_entities = [ent[0] for ent in ner_extraction_spacy(text)]

        # Combine both approaches
        combined_entities = combine_extractions(dict_entities, ner_entities)

        # Append structured results
        results.append({
            "text": text,
            "labels": snippet["labels"],
            "extracted_entities": combined_entities
        })

    return results

# Example usage
snippets = [
    {"text": "I have an issue with my billing and I want a refund.", "labels": ["Billing", "Refund"]},
    {"text": "My internet is not working, can you help me fix it?", "labels": ["Technical Issue", "Internet"]},
    {"text": "I want to complain about the service I received.", "labels": ["Complaint"]}
]

domain_knowledge = load_json("domain_knowledge.json")
structured_results = output_entities(snippets, domain_knowledge)

print("Structured Results:")
print(json.dumps(structured_results, indent=4))
```

```python
from flask import Flask, request, jsonify
import spacy
import json
import re

# Load spaCy model for NER
nlp = spacy.load("en_core_web_sm")

# Sample domain knowledge for dictionary lookup
domain_knowledge = {
    "competitors": ["BrandX", "BrandY", "BrandZ"],
    "features": ["internet", "support", "pricing"],
    "pricing": ["affordable", "cheap", "expensive"]
}

# Initialize Flask app
app = Flask(__name__)

# Function for dictionary-based entity lookup
def dictionary_lookup(text, domain_knowledge):
    entities = {"competitors": [], "features": [], "pricing": []}
    
    for competitor in domain_knowledge["competitors"]:
        if re.search(r"\b" + re.escape(competitor) + r"\b", text, re.IGNORECASE):
            entities["competitors"].append(competitor)
    
    for feature in domain_knowledge["features"]:
        if re.search(r"\b" + re.escape(feature) + r"\b", text, re.IGNORECASE):
            entities["features"].append(feature)
    
    for price in domain_knowledge["pricing"]:
        if re.search(r"\b" + re.escape(price) + r"\b", text, re.IGNORECASE):
            entities["pricing"].append(price)

    return entities

# Function for NER extraction using spaCy
def ner_extraction_spacy(text):
    doc = nlp(text)
    ner_entities = {"competitors": [], "features": [], "pricing": []}
    
    for ent in doc.ents:
        if ent.label_ == "ORG":
            ner_entities["competitors"].append(ent.text)
        elif ent.label_ == "MONEY":
            ner_entities["pricing"].append(ent.text)
        else:
            ner_entities["features"].append(ent.text)
    
    return ner_entities

# Function for combining both dictionary lookup and NER
def combine_extractions(dict_entities, ner_entities):
    final_entities = {
        "competitors": list(set(dict_entities["competitors"] + ner_entities["competitors"])),
        "features": list(set(dict_entities["features"] + ner_entities["features"])),
        "pricing": list(set(dict_entities["pricing"] + ner_entities["pricing"]))
    }
    return final_entities

# Function for text summarization (basic implementation)
def summarize_text(text):
    # Simple 1-2 sentence summary (just returning first 1-2 sentences as summary)
    sentences = text.split('.')
    return '. '.join(sentences[:2]) + '.' if len(sentences) > 1 else sentences[0] + '.'

# Route for accepting text and returning predictions, extracted entities, and summary
@app.route('/process', methods=['POST'])
def process_text():
    data = request.get_json()  # Get the JSON data from the request
    text = data.get('text', '')  # Extract the text from the JSON

    # Step 1: Perform dictionary lookup and NER extraction
    dict_entities = dictionary_lookup(text, domain_knowledge)
    ner_entities = ner_extraction_spacy(text)
    
    # Combine results from both approaches
    final_entities = combine_extractions(dict_entities, ner_entities)
    
    # Step 2: Predict labels (for simplicity, we return dummy labels for now)
    predicted_labels = ["Technical Issue", "Complaint"]  # Example predicted labels
    
    # Step 3: Generate a summary
    summary = summarize_text(text)

    # Return the result in JSON format
    return jsonify({
        "predicted_labels": predicted_labels,
        "extracted_entities": final_entities,
        "summary": summary
    })

# Run the Flask app
if __name__ == "__main__":
    app.run(debug=True, host='0.0.0.0', port=5000)
```
